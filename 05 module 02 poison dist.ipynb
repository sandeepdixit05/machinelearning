```python
# If the average number of calls to a call center is 10 per hour, what is the Poisson parameter (λ) over 3 hours?
# The Poisson parameter (λ) is the average number of events in a given time period.
def calculate_poisson_parameter(average_calls_per_hour, hours):
    return average_calls_per_hour * hours
# Example usage:
average_calls_per_hour = 10
hours = 3
poisson_parameter = calculate_poisson_parameter(average_calls_per_hour, hours)
print(f"The Poisson parameter (λ) over {hours} hours is: {poisson_parameter}")


```

```python
from scipy.stats import poisson
import numpy as np

# Parameters
lambda_rate = 10
k = 7

# Calculate Poisson probability
prob = poisson.pmf(k, lambda_rate)
print(f"Probability of getting exactly 7 calls in one hour: {prob:.4f}")

x = np.arange(0, 21)
y = poisson.pmf(x, lambda_rate)
import matplotlib.pyplot as plt
plt.bar(x, y, width=0.8, color='blue', alpha=0.7)
plt.title('Poisson Distribution PMF (λ = 10)')
plt.xlabel('Number of Calls')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import poisson

# Values from 0 to 20
x = np.arange(0, 21)
cdf = poisson.cdf(x, 10)

plt.figure(figsize=(8, 5))
plt.step(x, cdf, where='post', color='darkgreen', label='CDF')
plt.scatter(x, cdf, color='green')
plt.title('Poisson CDF (λ = 10)')
plt.xlabel('Number of Calls in 1 Hour')
plt.ylabel('Cumulative Probability P(X ≤ x)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
