```python
import matplotlib.pyplot as plt
import seaborn as sn

# Plot a bar chart
def plot_bar_chart(data, x_col, y_col):
    try:
        plt.figure(figsize=(10, 6))
        sn.barplot(x=x_col, y=y_col, data=data)
        plt.title(f'{y_col} by {x_col}')
        plt.xlabel(x_col)
        plt.ylabel(y_col)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"An error occurred while plotting the bar chart: {e}")

# Main function to run the script
file_path = 'IPL_IMB381IPL2013.csv'
import pandas as pd
df = pd.read_csv(file_path)

if df is not None:
    print(df.head())  # Display the first few rows of the dataframe
    plot_bar_chart(df, 'TEAM', 'T-RUNS')  # Example columns, adjust as necessary


```
