```python
# Binomial Distribution
# https://en.wikipedia.org/wiki/Binomial_distribution
# import libraries required for this example
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import binom

# set the parameters for the binomial distribution
n = 20  # number of trials
p = 0.1  # probability of success
s = 5 # number of samples
# Find the probablity of getting k successes in n trials
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)
# Create a DataFrame for plotting
df = pd.DataFrame({'x': x, 'pmf': pmf})
print("DataFrame:\n", df)

plt.figure(figsize=(20, 6))
plt.title(f'Binomial Distribution PMF (n={n}, p={p})')
plt.xlabel('Number of Customer Will Return the Item (k)')
plt.ylabel('Probability Mass Function (PMF)')
plt.xticks(x)
plt.ylim(0, max(pmf) * 1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
# Add values as text on top of the bars
for i in range(len(df)):
    plt.text(df.x[i], df.pmf[i] + 0.01, f'{df.pmf[i]:.3f}', ha='center', va='bottom')

sns.barplot(x=df.x, y=df.pmf, data=df, color='blue', alpha=0.7)
plt.show()




cdf = binom.cdf(x, n, p)
# Create a DataFrame for CDF plotting, restrict cdf to 3 decimal places
df_cdf = pd.DataFrame({'x': x, 'cdf': cdf, '1-cdf': 1 - cdf})
df_cdf['cdf'] = df_cdf['cdf'].round(5)
df_cdf['1-cdf'] = df_cdf['1-cdf'].round(5)
print("CDF DataFrame:\n", df_cdf)


# Calcuate average and variance and standard deviation
mean = binom.mean(n, p)
variance = binom.var(n, p)
std_dev = np.sqrt(variance)
print(f"Mean: {mean}, Variance: {variance}, Standard Deviation: {std_dev}")

```
